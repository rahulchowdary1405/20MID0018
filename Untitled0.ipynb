```python
#import uniform distribution
from scipy.stats import uniform
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def t_plot(data, fun_type):
    ax = sns.histplot(data, binrange=[1, plot_scale - 1],
                      bins=plot_scale - 1,
                      kde=True,
                      stat="density",
                      kde_kws=dict(cut=cut_scale),
                      alpha=.4,
                      edgecolor=(0.25, 0.25, 1, .6),
                      label='Uniform Distribution',
                      discrete=True)
    plt.title(fun_type)
    plt.show()

# 1st case
size = 200
plot_scale = 10
cut_scale = plot_scale / np.log10(plot_scale)
data = uniform.rvs(size=size, scale=plot_scale)
fun_type = "Uniform Discrete (Size = 200, Scale = 10)"
t_plot(data, fun_type)

# 2nd case
size = 500
plot_scale = 20
cut_scale = plot_scale / np.log10(plot_scale)
data = uniform.rvs(size=size, scale=plot_scale)
fun_type = "Uniform Discrete (Size = 500, Scale = 20)"
t_plot(data, fun_type)

# 3rd case
size = 50
plot_scale = 5
cut_scale = plot_scale / np.log10(plot_scale)
data = uniform.rvs(size=size, scale=plot_scale)
fun_type = "Uniform Discrete (Size = 50, Scale = 5)"
t_plot(data, fun_type)

# 4th case
size = 1000
plot_scale = 50
cut_scale = plot_scale / np.log10(plot_scale)
data = uniform.rvs(size=size, scale=plot_scale)
fun_type = "Uniform Discrete (Size = 1000, Scale = 50)"
t_plot(data, fun_type)

# 5th case
size = 25
plot_scale = 3
cut_scale = plot_scale / np.log10(plot_scale)
data = uniform.rvs(size=size, scale=plot_scale)
fun_type = "Uniform Discrete (Size = 25, Scale = 3)"
t_plot(data, fun_type)
```

Observations:

Size = 200, Scale = 10: The distribution shows a uniform spread across the range of 1 to 10. It appears fairly smooth and evenly distributed.
Size = 500, Scale = 20: Increasing the size and scale results in a wider, flatter distribution still spanning the full range. It's more spread out compared to the first case.
Size = 50, Scale = 5: Reducing the size and scale produces a more peaked distribution, concentrated in the middle compared to the previous cases, but still maintaining a uniform shape.
Size = 1000, Scale = 50: Further increasing the size and scale creates an even wider, flatter distribution that still conforms to the uniform shape. It's the most spread out of the cases.
Size = 25, Scale = 3: Decreasing the size and scale produces a more concentrated distribution, but it still exhibits uniform characteristics. It's the most peaked of the cases.

```python
from scipy.stats import binom

# 1st case
data_binom = binom.rvs(n=100, p=0.2, loc=0, size=1000)
ax = sns.distplot(data_binom, kde=False, hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Binomial', ylabel='Frequency')

# 2nd case
data_binom = binom.rvs(n=60, p=0.7, loc=0, size=1000)
ax = sns.distplot(data_binom, kde=False, hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Binomial', ylabel='Frequency')

# 3rd case
data_binom = binom.rvs(n=15, p=0.95, loc=0, size=1000)
ax = sns.distplot(data_binom, kde=False, hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Binomial', ylabel='Frequency')

# 4th case
data_binom = binom.rvs(n=40, p=0.3, loc=0, size=1000)
ax = sns.distplot(data_binom, kde=False, hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Binomial', ylabel='Frequency')

# 5th case
data_binom = binom.rvs(n=20, p=0.8, loc=0, size=1000)
ax = sns.distplot(data_binom, kde=False, hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Binomial', ylabel='Frequency')
```

OBSERVATIONS OF BINOMIAL DISTRIBUTION
• n=100, p=0.2, loc=0, size=1000: This distribution displays the typical bell-shaped curve of a binomial distribution with a large number of trials (n=100) and a relatively low success probability (p=0.2). The distribution is symmetric and centered around its mean, indicating that outcomes are evenly spread around the expected value.
• n=60, p=0.7, loc=0, size=1000: Increasing the number of trials to 60 while raising the success probability to 0.7 results in a narrower and taller binomial distribution compared to the first case. The distribution is still symmetric, but its peak is higher and narrower, suggesting a higher likelihood of success within a smaller range of outcomes.
• n=15, p=0.95, loc=0, size=1000: Reducing the number of trials to 15 and significantly increasing the success probability to 0.95 produces a heavily skewed distribution. The distribution is highly concentrated towards the higher values, with a long tail extending towards lower values. This indicates a high likelihood of success with very few trials.
• n=40, p=0.3, loc=0, size=1000: Adjusting the parameters to n=40 and p=0.3 results in a symmetric binomial distribution similar to the first case but with a different combination of parameters. The distribution is centered around its mean and has a moderate spread, reflecting a balanced likelihood of success across the trials.
• n=20, p=0.8, loc=0, size=1000: Similar to the first case, this distribution also exhibits a bell-shaped curve. With n=20 trials and a relatively high success probability of 0.8, the distribution is symmetric and centered around its mean, indicating a high likelihood of success with a moderate number of trials.

```python
from scipy.stats import poisson
import seaborn as sns

# 1st case
data_poisson_1 = poisson.rvs(mu=8, size=10000)
ax = sns.distplot(data_poisson_1,
                 kde=False,
                 hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')

# 2nd case
data_poisson_2 = poisson.rvs(mu=15, size=10000)
ax = sns.distplot(data_poisson_2,
                 kde=False,
                 hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')

# 3rd case
data_poisson_3 = poisson.rvs(mu=4, size=10000)
ax = sns.distplot(data_poisson_3,
                 kde=False,
                 hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')

# 4th case
data_poisson_4 = poisson.rvs(mu=9, size=10000)
ax = sns.distplot(data_poisson_4,
                 kde=False,
                 hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')

# 5th case
data_poisson_5 = poisson.rvs(mu=6, size=20000)
ax = sns.distplot(data_poisson_5,
                 kde=False,
                 hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')
```

OBSERVATIONS OF POISSON DISTRIBUTION
• mu=8, size=10000: This Poisson distribution with a mean of 8 exhibits the typical single-peaked shape. The distribution is skewed to the right, as expected for a Poisson distribution.
• mu=15, size=10000: Increasing the mean to 15 results in a wider, more spread-out Poisson distribution. The distribution is still single-peaked but has a larger spread compared to the first case.
• mu=4, size=10000: Decreasing the mean to 4 produces a Poisson distribution that is more heavily skewed towards the lower values. The distribution is more concentrated on the left side.
• mu=9, size=10000: Using a mean of 9 creates a Poisson distribution that is more symmetric than the previous examples. The distribution is still skewed to the right but is more centered around the mean.
• mu=6, size=20000: Finally, reducing the mean to 6 but increasing the sample size to 20,000 yields a Poisson distribution that is still single-peaked but more smoothed out. The distribution is less skewed compared to the third case.

```python
from scipy.stats import geom
import seaborn as sns
import numpy as np

# 1st case
data_geom = geom.rvs(p=0.4,
                     size=1500,
                     random_state=21)

ax = sns.distplot(data_geom,
                  kde=False,
                  hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Geometric Distribution', ylabel='Frequency')

# 2nd case
data_geom = geom.rvs(p=0.7,
                     size=1500,
                     random_state=21)

ax = sns.distplot(data_geom,
                  kde=False,
                  hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Geometric Distribution', ylabel='Frequency')

# 3rd case
data_geom = geom.rvs(p=0.2,
                     size=1500,
                     random_state=21)

ax = sns.distplot(data_geom,
                  kde=False,
                  hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Geometric Distribution', ylabel='Frequency')

# 4th case
data_geom = geom.rvs(p=0.5,
                     size=2000,
                     random_state=21)

ax = sns.distplot(data_geom,
                  kde=False,
                  hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Geometric Distribution', ylabel='Frequency')

# 5th case
data_geom = geom.rvs(p=0.9,
                     size=700,
                     random_state=21)

ax = sns.distplot(data_geom,
                  kde=False,
                  hist_kws={"linewidth": 25, "alpha": 1})
ax.set(xlabel='Geometric Distribution', ylabel='Frequency')
```

p=0.4, size=1500: This geometric distribution with a success probability of 0.4 exhibits the characteristic exponential-like decay. The distribution is moderately skewed towards the lower values.

p=0.7, size=1500: Increasing the success probability to 0.7 results in a geometric distribution that is more concentrated towards the lower values. The distribution is even more heavily skewed compared to the first case.

p=0.2, size=1500: Decreasing the success probability to 0.2 produces a geometric distribution that is more spread out and has a longer tail. The distribution is less skewed and more evenly distributed.

p=0.5, size=2000: Using a success probability of 0.5 and increasing the sample size to 2000 creates a geometric distribution that is still exponentially decaying but more smoothed out. The distribution is less jagged compared to the previous cases.

p=0.9, size=700: Finally, increasing the success probability to 0.9 while reducing the sample size to 700 yields a geometric distribution that is heavily skewed towards the lower values. The distribution is the most concentrated of the geometric cases.
